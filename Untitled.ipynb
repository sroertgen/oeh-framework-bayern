```python
def get_item_title(soup, subject, context, grade):
    global row_i
    for item in soup.find_all('span', 'head-absatz-title-short'):
        fullStatement = item.get_text().strip()
        humanCodingScheme = context + '_' + subject + str(grade)
        listEnumeration = row_i
        row_a = 'A' + str(row_i)
        row_b = 'B' + str(row_i)
        row_c = 'C' + str(row_i)
        worksheet.write(row_a, fullStatement)
        worksheet.write(row_b, humanCodingScheme)
        worksheet.write(row_c, listEnumeration)
        row_i += 1
        print(f'row is: {row_i}')
        print(fullStatement)
```

```python
row_i = 1
for i in range(1, 14):
    for subject in subjects:
        for context in educationalContext:
            url = 'https://www.lehrplanplus.bayern.de/fachlehrplan/' + context + '/' + str(i) + '/' + subject
            print(url)
            r = requests.get(url)
            soup = BeautifulSoup(r.text, 'html.parser')
            get_item_title(soup, context, subject, i)
workbook.close()
```
